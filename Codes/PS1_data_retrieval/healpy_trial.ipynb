```python
import numpy as np
import healpy as hp
import matplotlib.pyplot as plt
```

```python
data = np.arange(10)
theta = np.radians(np.arange(10, 110, 10))
phi = np.radians(np.linspace(0 , 100, 10))

nside = 8
print(np.degrees(hp.nside2resol(nside)))

pixel_indices = hp.ang2pix(nside, theta, phi)
m = np.zeros(hp.nside2npix(nside))
m[pixel_indices] = data
hp.mollview(m)
```

```python
n = np.zeros(hp.nside2npix(nside))
pixel_index = hp.ang2pix(nside, np.pi/2 , 2* np.pi-0.001)
n[pixel_index] = 1
hp.mollview(n)
print(np.shape(n))
```
